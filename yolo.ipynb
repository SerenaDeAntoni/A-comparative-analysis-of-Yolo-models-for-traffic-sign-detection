## Dataset preprocessing

```python
#coping all the element of a folder in an another folder

import shutil
import os

src = '/content/drive/MyDrive/computervision/dataset/dataset/datasets/train/images'
dst = '/content/drive/MyDrive/computervision/dataset/dataset/datasets/train_aug/images'

for file_name in os.listdir(src):
    full_file_name = os.path.join(src, file_name)
    if os.path.isfile(full_file_name) :
        shutil.copy(full_file_name, dst)
```

```python
# change all the names of the .txt files inside a folder with currentname+_0
import os
import sys
import re
import shutil

folder = '/content/drive/MyDrive/computervision/dataset/dataset/datasets/train_aug/labels'
for filename in os.listdir(folder):
    if filename.endswith(".txt"):
        newname = re.sub(r'\.txt$', '_0.txt', filename)
        shutil.move(os.path.join(folder, filename), os.path.join(folder, newname))
```

```python
# Function to perform data augmentation

import cv2
import os
import numpy as np

def augment_data(input_dir, output_dir, num_augmented_images_per_original):
    # Create output directory if it doesn't exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Loop through each image in the input directory
    for filename in os.listdir(input_dir):
        if filename.endswith(".jpg") or filename.endswith(".jpeg"):
            # Read the image
            image = cv2.imread(os.path.join(input_dir, filename))

            # Apply data augmentation techniques and save the augmented images
            for i in range(num_augmented_images_per_original):
                augmented_image = image.copy()

                alpha = np.random.uniform(0.5, 1.5)  # Randomly adjust contrast (0.5 to 1.5)
                beta = np.random.randint(-50, 50)     # Randomly adjust brightness (-50 to 50)
                augmented_image = cv2.convertScaleAbs(augmented_image, alpha=alpha, beta=beta)

                # Save the augmented image
                output_filename = f"{os.path.splitext(filename)[0]}_0.jpg"
                cv2.imwrite(os.path.join(output_dir, output_filename), augmented_image)

# Example usage
input_directory = "/content/drive/MyDrive/computervision/dataset/dataset/datasets/train/images"
output_directory = "/content/drive/MyDrive/computervision/dataset/dataset/datasets/train_aug/images"
num_augmented_images_per_original = 1  # Number of augmented images to generate per original image

# Perform data augmentation
augment_data(input_directory, output_directory, num_augmented_images_per_original)
```
